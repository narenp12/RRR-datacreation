```python
import pandas as pd
import numpy as np
import plotly.express as px
import sklearn as sk
import nltk
import gensim
import spacy
import sklearn as sk
from keybert import KeyBERT
import hf_xet
```

```python
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
import torch
```

```python
docs = pd.read_csv('/kaggle/input/rrr-docs/paper_doc_data.csv')
```

```python
kw_model = KeyBERT(model = 'all-MiniLM-L6-v2')
docs["keyphrases"] = docs["summary"].apply(lambda x: kw_model.extract_keywords(x, top_n=5, stop_words='english'))
docs["concepts_found"] = docs["keyphrases"].apply(lambda x: [kw for kw, _ in x])
```

```python
docs.to_csv("newdocs.csv")
```
